```python
from aimacode.logic import PropKB
from aimacode.planning import Action
from aimacode.search import (
    Node, Problem,
)
from aimacode.utils import expr
from lp_utils import (
    FluentState, encode_state, decode_state,
)
from my_planning_graph import PlanningGraph
from itertools import starmap
from functools import lru_cache
from run_search import PROBLEMS, SEARCHES, run_search, show_solution, PrintableProblem
import pandas as pd
import numpy as np
from timeit import default_timer as timer
from plotnine import *
from tabulate import tabulate
import logging
```

```python
logger = logging.Logger
```

```python
def run_search_dict(problem, pname, search_function, 
                  sname, parameter, parameter_name):
    start = timer()
    ip = PrintableProblem(problem)
    if parameter is not None:
        node = search_function(ip, parameter)
    else:
        node = search_function(ip)
    end = timer()
    print("\nExpansions   Goal Tests   New Nodes")
    print("{}\n".format(ip))
    show_solution(node, end - start)
    solution = ["{}{}".format(action.name, action.args) for action in node.solution()]
           
    return {
        'problem': pname,
        'search': sname,
        'parameter': parameter_name,
        'plan_length': len(node.solution()),
        'expansions': ip.succs,
        'goal_tests': ip.goal_tests,
        'new_nodes': ip.states,
        'elapsed_time': end - start,
        'solution': solution
    }
```

```python
%%time
problems = PROBLEMS

searches = [x for x in SEARCHES if x[0] in ['breadth_first_search',
                                'depth_first_graph_search',
                                'uniform_cost_search',
                                'astar_search']]
results = []
for pname, p in problems:
    for sname, s, h in searches:
        hstring = h if not h else " with {}".format(h)
        print("\nSolving {} using {}{}...".format(pname, sname, hstring))
        _p = p()
        _h = None if not h else getattr(_p, h)
        results.append(run_search_dict(_p, pname, s, sname, _h, h))
```

```python
df_results = pd.DataFrame(results)
df_results['repeated_action_pct'] = (
    1 - (df_results['solution'].map(lambda x: len(set(x)))
         / df_results['solution'].map(len)))
df_results['search_param'] = (df_results['search']
                              .str.replace('_search', '')
                               .str.cat(df_results['parameter'],
                                        sep='\n'))
```

```python
df_results.to_csv('planning_results.csv')
```

```python
def facet_plot(df, y, format_string=None):
    return (ggplot(df_results, aes( y=y, x='search_param',
                    ymax='{} * 1.2'.format(y))) 
     + geom_col(position='dodge')
     + facet_wrap('problem', ncol=1, scales='free')
     + theme(figure_size=[10, 4])
     + geom_text(aes(label=y), va='bottom', 
                 format_string=format_string))
```

```python
def save_plot(plot, name):
    plot.save('{}.png'.format(name), 
              path='/Users/mdiaz/CompSci/AI/aind/AIND-Planning/')
```

```python
def facet_save(y, format_string=None):
    plot = facet_plot(df_results, y, format_string)
    save_plot(plot, y)
    return plot
```
